```python
import numpy as np  
import matplotlib.pyplot as plt  
import pandas as pd  
from apyori import apriori 
```

```python
df = pd.read_csv('./clean2.xls')
df.head()
```

```python
df = df[["SEX","EDUCATION","MARRIAGE","default.payment.next.month"]]
df.shape
```

```python
records = []  
for i in range(0,30000):  
    records.append([str(df.values[i,j]) for j in range(0, 4)])
```

```python
records
```

```python
association_rules = apriori(df, min_support=0.0005, min_confidence=0.01, min_lift=2, min_length=2)  
association_results = list(association_rules)
association_results
```

```python
for item in association_results:

    # first index of the inner list
    # Contains base item and add item
    pair = item[0] 
    items = [x for x in pair]
    print('Rules : ', items)
    #print("Rule: " + items[0] + " -> " + items[1])

    #second index of the inner list
    print("Support: " + str(item[1]))

    #third index of the list located at 0th
    #of the third index of the inner list

    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```
