```python
import pandas as pd
df = pd.read_csv('./clean2.xls')
df.head()
```

```python
trainRow = df.iloc[1:20000,:]
testRow = df.iloc[20001:30000,:]
features = trainRow[["LIMIT_BAL","Pay_mean","bill_mean"]]
label = trainRow[["default.payment.next.month"]]
testFeatures = testRow[["LIMIT_BAL","Pay_mean","bill_mean"]]
testLabel = testRow[["default.payment.next.month"]]
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=1000)
model.fit(features,label)
predicted= model.predict(testFeatures) 
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(testLabel, predicted))
```
